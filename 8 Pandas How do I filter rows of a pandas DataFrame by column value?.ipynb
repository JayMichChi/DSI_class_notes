```python
# we want to find all the movies whose duration exceeds a certain amount of time....

#boolean: going to create a python list of booleans if the duration is True

#1 Need to build a "for loop"

booleans=[]
for genre in movies.genre:
    if genre == "Crime" or genre == "Action":
        booleans.append(True)
 
    else:
        booleans.append(False)
```
